```python
import torch
import matplotlib.pyplot as plt
```

```python
a = torch.tensor(torch.arange(-10.0, 11.0, 1.0),requires_grad=True)
b = a**2
plt.plot(a.detach(), b.detach());
```

```python
out = torch.sum(b)
out.backward()
print(a.grad)
print(a)
```

```python
plt.plot(a.detach(), a.grad.detach());
```

```python
# Gradient of a function
x = torch.tensor(7.0, requires_grad=True)
f = (x**2)+3
f.backward()
print(x.grad)
```

```python
from torch.utils.tensorboard import SummaryWriter
writer = SummaryWriter()
```

```python
lr = 0.1
x = torch.tensor(10.0, requires_grad=True)
for epoch in range(100):

    if x.grad is not None:
        x.grad.zero_()

    f = (x**2)+3
    f.backward()
    x.data = x.data - lr*x.grad
    writer.add_scalar("x", x.grad,epoch)
    print(x)
    print(x.grad)

    if x.grad <= 0.01:
        break

print(x)
print(f)
print(x.grad)

writer.flush()
writer.close() 
```
